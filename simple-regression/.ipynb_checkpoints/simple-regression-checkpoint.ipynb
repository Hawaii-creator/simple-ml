# ①データ読み込み

```python
import warnings
warnings.simplefilter('ignore')

from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
 
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['y'] = boston.target
df.head()
```

```python
df.shape
```

## ②説明変数、目的変数の設定

```python
X = df['RM'].values
y = df['y'].values
```

```python
n = len(X)
n
```

# ③単回帰分析(1)sklearn

```python
X_sk = X.reshape(-1, 1)
y_sk = y.reshape(-1, 1)
```

```python
model = linear_model.LinearRegression()
model.fit(X_sk, y_sk)
```

```python
#傾き
model.coef_
```

```python
#切片
model.intercept_
```

```python
#決定係数
r2_sk = model.score(X_sk, y_sk)
```

```python
#自由度調整済み決定係数
r2_adjust_sk = 1 - (1 - r2_sk) * (len(y) - 1) / (len(y) - 1 - 1)
```

```python
r2_adjust_sk
```

# ③単回帰分析(2)numpy

```python
#傾き
slope = ((np.dot(X,y) - y.sum() * X.sum() / n) / ((X ** 2).sum() - X.sum() ** 2 / n))
#切片
intercept = (y.sum() - slope * X.sum()) / n
#予測値
pred = slope * X + intercept
```

```python
#傾きと切片　(1)のsklearnと結果が一致することを確認
print("slope:", slope, "intercept:", intercept)
```

```python
#決定係数　(1)のsklearnと結果が一致することを確認
r2_bunbo = ((y - y.mean())**2).sum()
r2_bunshi = ((y - pred)**2).sum()
r2 = 1 - r2_bunshi / r2_bunbo
r2
```

```python
#自由度調整済み決定係数　(1)のsklearnと結果が一致することを確認
r2_adjust = 1 - (1 - r2) * (len(y) - 1) / (len(y) - 1 - 1)
r2_adjust
```
